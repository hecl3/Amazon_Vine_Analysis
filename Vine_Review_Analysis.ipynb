```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df
```

```python
#Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

vine_review_df = vine_df.loc[vine_df["total_votes"] >= 20]
vine_review_df.head()
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the 
#number of helpful_votes divided by total_votes is equal to or greater than 50%.

vine_helpful_df = vine_review_df.loc[vine_review_df["helpful_votes"]/vine_review_df["total_votes"] >= 0.5]
vine_helpful_df.head()
```

```python
#Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review
#was written as part of the Vine program (paid), vine == 'Y'

vine_paid_df = vine_helpful_df.loc[vine_helpful_df["vine"] == 'Y']
vine_paid_df.head()
```

```python
#Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

vine_unpaid_df = vine_helpful_df.loc[vine_helpful_df["vine"] == 'N']
vine_unpaid_df.head()
```

```python
#Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of 
#review (paid vs unpaid).

paid_five_star_number = vine_paid_df.loc[(vine_paid_df['star_rating']== 5)]["star_rating"].count()
paid_number = vine_paid_df["star_rating"].count()
percentage_five_star_vine = paid_five_star_number / paid_number * 100
print(f"5 Star Rating: {paid_five_star_number}")
print(f"Total Rating: {paid_number}")
print(f"Percentage of 5 star: {round(percentage_five_star_vine,2)}%")
```

```python
unpaid_five_star_number = vine_unpaid_df.loc[(vine_unpaid_df['star_rating']== 5)]["star_rating"].count()
unpaid_number = vine_unpaid_df["star_rating"].count()
percentage_five_star_vine = unpaid_five_star_number / unpaid_number * 100
print(f"5 Star Rating: {unpaid_five_star_number}")
print(f"Total Rating: {unpaid_number}")
print(f"Percentage of 5 star: {round(percentage_five_star_vine,2)}%")
```
